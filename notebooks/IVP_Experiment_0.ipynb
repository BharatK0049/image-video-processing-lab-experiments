### Importing Necessary Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Loading image

#### BGR being applied when using openCV


```python
img = cv2.imread("images.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img)
plt.axis("off")
plt.show()
```

Inference:
- OpenCV uses BGR format and reads image in that format, however matplotlib expects image in RGB format

#### Colormap being used when displaying image

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
plt.axis("off")
```

Inference:
- By default, Matplotlib applies a colormap when displaying 2D arrays.

#### Correct way to display grayscale image

```python
plt.imshow(gray, cmap="gray")
plt.axis("off")
```

Inference:
When grayscale colormap is mentioned, the image turns gray

#### What if I just try to apply grayscale to color image directly?


```python
plt.imshow(img, cmap = 'gray')
plt.axis('off')
```

Inference -
When image is already colored, matplotlib ignores colormap and displays it as is

#### Displaying BGR to RGB

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis("off")
```

### use of cvtcolor

cvtColor = convert color space

It applies a deterministic mathematical transformation to each pixel.

Personal note: use plt.axis('off') to remove x and y axis of the image
